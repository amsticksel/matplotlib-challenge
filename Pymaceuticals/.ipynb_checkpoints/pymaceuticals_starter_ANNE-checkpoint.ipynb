```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load, encoding="ISO-8859-1")
trial_df = pd.read_csv(clinical_trial_data_to_load, encoding="ISO-8859-1")
# Combine the data into a single dataset
mergedmouse = pd.merge(mouse_df, trial_df, on="Mouse ID")

# Display the data table for preview
mergedmouse
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mergedmousedrug = mergedmouse.groupby(["Drug", "Timepoint"], as_index=False).mean()
# Convert to DataFrame
# Preview DataFrame
mergedmousedrug
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

errormouse = mergedmouse.groupby(["Drug", "Timepoint"]).sem()

# Convert to DataFrame
emouse = errormouse.reset_index()
emouse.head()
#clean up the data a bit
remouse = emouse.drop(['Metastatic Sites', 'Mouse ID'], axis = 1)
remouse = remouse.rename(columns={'Tumor Volume (mm3)': 'Tumor Error'})
remouse

```

```python
sizemouse = mergedmouse.groupby(["Drug", "Timepoint"], as_index=False).mean()
sizemouse.reset_index()
# Convert to DataFrame

plainsizemouse = sizemouse.drop(['Metastatic Sites'], axis = 1)
pmouse = plainsizemouse.rename(columns={'Tumor Volume (mm3)': 'Tumor Size'})

#pivot the table so timepoint and drug are on the top with tumor size as the values
pivmouse = pmouse.pivot(index='Timepoint', columns='Drug', values='Tumor Size')
#find the margin of error
error = pivmouse.sem()
#create the plot
plt.figure(figsize=(8,6))
#plot each point with margin of error
plt.errorbar(pivmouse.index, pivmouse["Capomulin"], yerr=error["Capomulin"], color="firebrick", marker="x")
plt.errorbar(pivmouse.index, pivmouse["Ceftamin"], yerr=error["Ceftamin"], color="coral", marker="+")
plt.errorbar(pivmouse.index, pivmouse["Infubinol"], yerr=error["Infubinol"], color="olivedrab", marker="h")
plt.errorbar(pivmouse.index, pivmouse["Ketapril"], yerr=error["Ketapril"], color="cadetblue", marker="s")
plt.errorbar(pivmouse.index, pivmouse["Naftisol"], yerr=error["Naftisol"], color="deepskyblue", marker=".")
plt.errorbar(pivmouse.index, pivmouse["Placebo"], yerr=error["Placebo"], color="slateblue", marker="x")
plt.errorbar(pivmouse.index, pivmouse["Propriva"], yerr=error["Propriva"], color="darkorchid", marker="+")
plt.errorbar(pivmouse.index, pivmouse["Ramicane"], yerr=error["Ramicane"], color="darkseagreen", marker="+")
plt.errorbar(pivmouse.index, pivmouse["Stelasyn"], yerr=error["Stelasyn"], color="mediumvioletred", marker="h")
plt.errorbar(pivmouse.index, pivmouse["Zoniferol"], yerr=error["Zoniferol"], color="black", marker="s")
#format the plot
plt.title("Tumor Response to Drug Treatment in Mice")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=1)
plt.grid(True)
plt.savefig('TumorResponse.png')
plt.show()
```

## Metastatic Response to Treatment

```python
mergedmousemeta = mergedmouse.groupby(["Drug", "Timepoint"], as_index=False).mean()
# Convert to DataFrame
# Preview DataFrame
mergedmousemeta
#find the standard error grouped by drug and timepoint
errormousemeta = mergedmouse.groupby(["Drug", "Timepoint"]).sem()

#create a dataframe and reset index 
emousemeta = errormousemeta.reset_index()

#clean up the data presentation
remousemeta = emousemeta.drop(['Tumor Volume (mm3)', 'Mouse ID'], axis = 1)
metamouse = mergedmouse.groupby(["Drug", "Timepoint"], as_index=False).mean()
metamouse.reset_index()
#clean up
plainmetamouse = metamouse.drop(['Tumor Volume (mm3)'], axis = 1)

#pivot the table so metastatic sites are the values
pivmetamouse = plainmetamouse.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
#find margin of error
metaerror = pivmetamouse.sem()
#display dataframe
pivmetamouse
```

```python
# Generate the Plot (with Error Bars)
plt.figure(figsize=(8,6))
#plot the information. I chose not to leave any of the drugs out
plt.errorbar(pivmetamouse.index, pivmetamouse["Capomulin"], yerr=metaerror["Capomulin"], color="firebrick", marker="x")
plt.errorbar(pivmetamouse.index, pivmetamouse["Ceftamin"], yerr=metaerror["Ceftamin"], color="coral", marker="+")
plt.errorbar(pivmetamouse.index, pivmetamouse["Infubinol"], yerr=metaerror["Infubinol"], color="olivedrab", marker="h")
plt.errorbar(pivmetamouse.index, pivmetamouse["Ketapril"], yerr=metaerror["Ketapril"], color="cadetblue", marker="s")
plt.errorbar(pivmetamouse.index, pivmetamouse["Naftisol"], yerr=metaerror["Naftisol"], color="deepskyblue", marker=".")
plt.errorbar(pivmetamouse.index, pivmetamouse["Placebo"], yerr=metaerror["Placebo"], color="slateblue", marker="x")
plt.errorbar(pivmetamouse.index, pivmetamouse["Propriva"], yerr=metaerror["Propriva"], color="darkorchid", marker="+")
plt.errorbar(pivmetamouse.index, pivmetamouse["Ramicane"], yerr=metaerror["Ramicane"], color="darkseagreen", marker="+")
plt.errorbar(pivmetamouse.index, pivmetamouse["Stelasyn"], yerr=metaerror["Stelasyn"], color="mediumvioletred", marker="h")
plt.errorbar(pivmetamouse.index, pivmetamouse["Zoniferol"], yerr=metaerror["Zoniferol"], color="black", marker="s")
#format the plot
plt.title("Metastatic Spread During Treatment")
plt.ylabel("Metastatic Sites")
plt.xlabel("Time (Days)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=1)
plt.grid(True)
plt.savefig('MetastaticSpread.png')
plt.show()
# Show the Figure
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mergedmice = mergedmouse.groupby(["Drug", "Timepoint"], as_index=False).count()
# Convert to DataFrame
# Preview DataFrame
mergedmice
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

mergedmice.reset_index()
# Convert to DataFrame
plainmice = mergedmice.drop(['Tumor Volume (mm3)', 'Metastatic Sites'], axis = 1)

#cleaning up the data a bit
pmice = plainmice.rename(columns={'Mouse ID': 'Mice'})

pivmice = pmice.pivot(index='Timepoint', columns='Drug', values='Mice')

pivmice
# Preview DataFrame
```

```python
# Generate the Plot (Accounting for percentages)
plt.figure(figsize=(8,6))
#there is probably a nicer way to do this but I could not figure it out
plt.errorbar(pivmice.index, ((pivmice["Capomulin"]/25)*100), color="firebrick", marker="x")
plt.errorbar(pivmice.index, ((pivmice["Ceftamin"]/25)*100), color="coral", marker="+")
plt.errorbar(pivmice.index, ((pivmice["Infubinol"]/25)*100), color="olivedrab", marker="h")
plt.errorbar(pivmice.index, ((pivmice["Ketapril"]/25)*100), color="cadetblue", marker="s")
plt.errorbar(pivmice.index, ((pivmice["Naftisol"]/25)*100), color="deepskyblue", marker=".")
plt.errorbar(pivmice.index, ((pivmice["Placebo"]/25)*100), color="slateblue", marker="x")
plt.errorbar(pivmice.index, ((pivmice["Propriva"]/26)*100), color="darkorchid", marker="+")
plt.errorbar(pivmice.index, ((pivmice["Ramicane"]/25)*100), color="darkseagreen", marker="+")
plt.errorbar(pivmice.index, ((pivmice["Stelasyn"]/26)*100), color="mediumvioletred", marker="h")
plt.errorbar(pivmice.index, ((pivmice["Zoniferol"]/25)*100), color="black", marker="s")


#formatting the plot
plt.title("Mice Survival Rate During Treatment")
plt.ylabel("Survival %")
plt.xlabel("Time (Days)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=1)
plt.grid(True)
# Show the Figure
plt.savefig('MiceSurvival.png')
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
sizemouse = mergedmouse.groupby(["Drug", "Timepoint"], as_index=False).mean()
sizemouse.reset_index()
# Convert to DataFrame

plainsizemouse = sizemouse.drop(['Metastatic Sites'], axis = 1)
pmouse = plainsizemouse.rename(columns={'Tumor Volume (mm3)': 'Tumor Size'})

#pivot the data on tumor size
pivmouse = pmouse.pivot(index='Timepoint', columns='Drug', values='Tumor Size')

# calculate the percent change for each item
perchange = (pivmouse.loc[45, :] - pivmouse.loc[0, :])/pivmouse.loc[0, :] * 100
perchange
```

```python
#add the information to a tuple
tumorgrowth = (perchange['Capomulin'],
              perchange['Ceftamin'],
              perchange['Infubinol'],
              perchange['Ketapril'],
              perchange['Naftisol'],
              perchange['Placebo'],
              perchange['Propriva'],
              perchange['Ramicane'],
              perchange['Stelasyn'],
               perchange['Zoniferol'],
              )
#create a container for drug labels
labels=[]
passes=[]
print (tumorgrowth)

#create a label for each drug with the percentage change value
for x in tumorgrowth:
        labels.append(x)
#create a label for the drug names
drugnames = ('Capomulin', 'Ceftamin', 'Infubinol', 'Ketapril', 'Naftisol', 'Placebo', 'Propriva', 'Ramicane', 'Setalsyn', 'Zoniferol')
y_pos = np.arange(len(drugnames))
#create a figure and plot it by color
plt.figure(figsize=(8,6))
plt.bar(y_pos, tumorgrowth, align='center', zorder=3, color=("c","r","r","r","r","r","r","c","r","r"), width=.75)
#format the graph
plt.xticks(y_pos, drugnames, rotation=90)
plt.xlabel('Drug Name')
plt.ylabel('% Change in Tumor Size')
plt.title('Tumor Growth Change Over 45-day Treatment')

plt.grid(zorder=0)
#create labels to go on the bars

for i in range(len(y_pos)):
    plt.text(x = y_pos[i]-.3 , y = tumorgrowth[i]+.5, s=(f'{round(labels[i])}%'))
plt.savefig('SummaryGraph.png')
plt.show()
```

### Observaions
* Capomulin and Ramicane were the only two drugs to show a decrease in tumor size.
* Stelasyn, Propriva, and Ifubinol all had greater fatality rates than the placebo.
* Capomulin had a slightly smaller margin of error than Ramicane for tumor size, but margin of error is very close for metastatic spread.
